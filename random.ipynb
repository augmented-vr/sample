```python
import random

# 1から6までのランダムな整数を作る
dice = random.randint(1, 6)

# 結果を表示する
print(f"サイコロの目は... {dice}")
```

```python
import random

# おみくじのリストを作成
omikuji = ["大吉", "中吉", "吉", "凶"]

# リストからランダムに一つ選ぶ
result = random.choice(omikuji)

# 結果を表示する
print(f"今日の運勢は...{result}")
```
