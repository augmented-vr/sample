```python
# 果物の名前が入ったリストを用意する
fruits = ["もも", "なし", "かき"]

# enumerate関数でインデックス番号と要素を同時に取得する
for index, fruit in enumerate(fruits):
  print(f"インデックス{index}のデータは{fruit}です。")
```

```python
# ランキングのリストを用意する
ranking = ["金メダル", "銀メダル", "銅メダル"]

# start=1 を指定して、1から始まる順位と要素を取得する
for rank, item in enumerate(ranking, start=1):
  print(f"{rank}位: {item}")
```
