```python
# Googleドライブをcolabに接続(マウント)するための機能を入れる
from google.colab import drive

# 自分のGoogleドライブを /content/drive という場所にマウントする
drive.mount("/content/drive")
```

```python
# Googleドライブの「マイドライブ」の中に「memo.txt」を作るパスを指定
file_path = "/content/drive/MyDrive/memo.txt"

# ファイルを書き込みモード("w")で開く
with open(file_path, "w") as file:
  # ファイルに文字を書き込む(\n は改行の記号)
  file.write("Googleドライブへの保存テストです。\n")
  file.write("Googleドライブに保存しました!")

print("ファイルへの書き込みが終わりました。")
```

```python
# 読み込むファイルのパスを指定
file_path = "/content/drive/MyDrive/memo.txt"

# ファイルを読み込みモード("r")で開く
with open(file_path, "r") as file:
  # ファイルの中身をすべてよみこんで変数に入れる
  content = file.read()

# 読み込んだ内容を表示する
print("読み込んだテキスト: ")
print(content)
```

```python
# 果物のリストを変数として定義する
favorite_fruits = ["もも", "なし", "かき"]

# 新しい保存先のパスを指定
list_file_path = "/content/drive/MyDrive/fruits.txt"

# ファイルを書き込みモード("w")で開く
with open(list_file_path, "w") as file:
  # for文を使って、リストの中身を順番に書き込む
  for fruit in favorite_fruits:
    file.write(fruit + "\n")

print("リストのデータをファイルに保存しました!")
```

```python
# 読み込むファイルのパスを指定
list_file_open = "/content/drive/MyDrive/fruits.txt"

# ファイルを読み込みモード("r")で開く
with open(list_file_open, "r") as file:
  # ファイルの中身を全て読み込む
  saved_fruits = file.read()

# 読み込んだ内容を表示する
print("好きな果物: ")
print(saved_fruits)
```
